```python
import numpy as np 
def compute_inverse(key):
  Inverse = np.linalg.inv(key)
  determinant = np.linalg.det(key)

  for i in range(1,27):
    deeterminant = determinant % 26
    if determinant * i % 26 == 1:
      Mul_Inv = i % 26

  Inverse = (Inverse*determinant*Mul_Inv)%26
  return Inverse




alphabets = [ chr(ord('a')+i) for i in range(26) ]

row = int(input("Enter the row of key: "))
key = input("Enter the key: ")

def compute_matrix_of_key(key):
  key_matrix = list()
  l = 0

  for j in range(row):
    k = 0
    u = list()
    for i in range(row):
      try:
        u.append(alphabets.index(key[l]))
      except:
        random = alphabets[k]
        u.append(alphabets.index(random))
        k = k+1
      l = l+1
    key_matrix.append(u)
  return key_matrix

key_matrix = np.array(compute_matrix_of_key(key))
# put key = udph for test 
key_inverse = compute_inverse(np.array(key_matrix))


#generate matrix of plain text
import math
text = input("Enter the text to be ciphered: ")
k = 0
text_matrix = list()
for j in range(math.ceil(len(text)/2)):
  u = list()
  for i in range(row):
    u.append(alphabets.index(text[k]))
    k = k + 1
  text_matrix.append(u)

#ciphering
def ciphering(text_matrix = text_matrix, key_matrix = key_matrix):
  cipher = list()
  for i in range(len(text_matrix)):
    product = np.multiply(key_matrix, np.array(text_matrix[i]))
    product = product.flatten()
    
    for i in range(0,len(product),row):
      product[i] = ( product[i] + product[i+1] ) % 26
      cipher.append(alphabets[product[i]])
  return ''.join(cipher)  

print("Ciphered text: ",ciphering() )



#deciphering

ciphered_matrix = compute_matrix_of_key(ciphering())

def deciphering(text_matrix = ciphered_matrix, key_matrix = key_inverse):
    decipher = list()
    for i in range(len(text_matrix)):
        product = np.multiply(key_matrix, np.array(ciphered_matrix[i]))
        product = product.flatten()
    
        for i in range(0,len(product),row):
            product[i] = ( product[i] + product[i+1] ) % 26
            
            decipher.append(alphabets[math.ceil(product[i])])
    return ''.join(decipher)  

print("Deciphered text is: ",deciphering())
```
